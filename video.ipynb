```python
import cv2

cap=cv2.VideoCapture(0)

width=int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height=int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

writer=cv2.VideoWriter('mysupervideo.mp4',cv2.VideoWriter_fourcc(*'DVIX'),20,(width,height))

while True:
    ret,frame=cap.read()
    
    #operations (drawing)
    
    
    
    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    writer.write(gray)
    cv2.imshow("frame",gray)
    
    if cv2.waitKey(1) & 0xFF==ord('q'):
        break
cap.release()
writer.release()
cv2.destroyAllWindows()
```

```python

import cv2
import time

cap=cv2.VideoCapture('mysupervideo.mp4')

if cap.isOpened==False:
    print("erroe file not found or wrong codec used")

while cap.isOpened():
    
    ret,frame=cap.read()
    
    if ret==True:
        
        #writer-->20 frames per sec
        time.sleep(1/20)
        cv2.imshow('frame',frame)
        
        if cv2.waitKey(10) & 0xFF==ord('q'):
            break
    else:
        break
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
cap=cv2.VideoCapture(0)

width=int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height=int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

#top left corner
x=width//2
y=height//2

#width and height pf rectangle 

#bottom right x+w , y+h

w=width//4
h=height//4


while True:
    ret,frame=cap.read()
    
    cv2.rectangle(frame,(x,y),(x+w,y+h),color=(0,255,0),thickness=5)
    
    cv2.imshow('frame',frame)
    
    if cv2.waitKey(1) & 0xFF==ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
        
```

```python
import cv2
cap=cv2.VideoCapture(0)

###CALL BACK FUNCTION 

def draw_rect(event,x,y,flags,param):
    global pt1,pt2,topleft_clicked,botRight_clicked
    
    if event ==cv2.EVENT_LBUTTONDOWN:
        
        #reset the rectangle (it checks if the rect there)
        
        if topleft_clicked == True and botRight_clicked==True:
            pt1=(0,0)
            pt2=(0,0)
            topleft_clicked=False
            botRight_clicked=False
        
        if topleft_clicked==False:
            pt1=(x,y)
            topleft_clicked=True
            
        elif botRight_clicked==False:
            pt2=(x,y)
            botRight_clicked=True
            
        

##GLOBAL VARIABLES
pt1=(0,0)
pt2=(0,0)
topleft_clicked=False
botRight_clicked=False

##CONNECT TO THE CALLBACK
cap=cv2.VideoCapture(0)

cv2.namedWindow('Test')
cv2.setMouseCallback('Test',draw_rect)







while True:
    ret,frame=cap.read()
    
    if topleft_clicked:
        cv2.circle(frame,center=pt1,radius=5,color=(0,0,255),thickness=-1)
    
    if topleft_clicked and botRight_clicked:
        cv2.rectangle(frame,pt2,pt2,(0,0,255),3)
    
    cv2.imshow('frame',frame)
    
    if cv2.waitKey(1) & 0xFF==ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
        
```
